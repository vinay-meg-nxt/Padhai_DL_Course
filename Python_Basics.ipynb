```python
# Program - While loop
i = 0
while i < 5 :
  print(i,i**3)
  i = i + 1
```

```python
# Use of functions
def sum_of_squares(stop_val,start_val=0):
  """Function which calculate sum of squares from start_val to stop_val"""
  sum = 0
  for i in range(start_val,stop_val):
    num = i+1
    sum += num*num
  return sum

# function call
num = 4
sum = sum_of_squares(num)
print("Sum of squares of numbers till",num, "is",sum)
```

#Python Lists

```python
# List of colours - same types - Homogeneous
colours = ["Black", "Blue","Red","Green", "White"]
type(colours)
len(colours)
colours.append("Brown")
other_colours = ["Grey","Purple"]
colours += other_colours
```

```python
# Access
print(colours[0])

# Print from index 0 to 2
print(colours[0:3])
```

## Shallow Copy

```python
copy_colors = colours
copy_colors[0] = "Yellow"
print(copy_colors)
print(colours)
```

## Deep Copy

```python
colours[0] = "Black"
copy_colors = list(colours)
copy_colors[0] = "Yellow"
print(copy_colors)
print(colours)
```

```python
print(colours)
```

## Other Utility Functions

```python
# List from range
numbers = list(range(1,5))
print(numbers)

# Use of 'map' function
squares = list(map(lambda x:x*x,numbers))
print(squares)

# Use of 'filter' function
even = list(filter(lambda x:x%2 == 0,numbers))
print(even)
```

```python
import time
my_tuple = tuple(range(100000))
my_list  = list(range(100000))

print(my_tuple[1:25])
print(my_list[1:25])

tic = time.time()
lambda x:x*x,my_list
toc = time.time()
print("Elapsed time in sec for list calculations",toc-tic)

tic = time.time()
lambda x:x*x,my_tuple
toc = time.time()
print("Elapsed time in sec for tuple calculations",toc-tic)

n1 = 5617
n2 = 673561
n3 = 5617123456

# Access time for list
tic = time.time()
print( n1 in my_list)
print( n2 in my_list)
print( n3 in my_list)
toc = time.time()
print("Elapsed time in sec for search in list is ",toc-tic)

# Access time for tuple
tic = time.time()
print(n1 in my_tuple)
print(n2 in my_tuple)
print(n3 in my_tuple)
toc = time.time()
print("Elapsed time in sec for search in tuple is ",toc-tic)
```

```python
#Sets are good for accessing, can't have duplicate values
my_set = set(my_list)

# Access time for tuple
tic = time.time()
print(n1 in my_set)
print(n2 in my_set)
print(n3 in my_set)
toc = time.time()
print("Elapsed time in sec for search in set is ",toc-tic)

my_set2 = {1,"go", True,5}
print(my_set2)
```
